```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('bmh')
x = np.linspace(0,20,200)
plt.title('sin & cosin figure')
plt.xlabel('x value')
plt.ylabel('y value')
plt.plot(x,np.sin(x),'--',c='red')
plt.legend(('cos(x)'))
plt.plot(x,np.cos(x),'-.',c='black')
plt.legend(('sin(x)'))
```

```python
plt.subplot(2,1,1)
plt.title('sin & cosin figure')
plt.xlabel('x value')
plt.ylabel('y value')
plt.plot(x,np.sin(x),'--',c='red')
plt.legend(('cos(x)'))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x),'-.',c='black')
plt.legend(('sin(x)'))
```

```python
fig = plt.figure()
plt.subplot(1,2,1)
plt.title('sin & cosin figure')
plt.xlabel('x value')
plt.ylabel('y value')
plt.plot(x,np.sin(x),'--',c='red')
plt.legend(('cos(x)'))
plt.subplot(1,2,2)
plt.plot(x,np.cos(x),'-.',c='black')
plt.legend(('sin(x)'))
fig.savefig('sin & cosin figure.png')
```
